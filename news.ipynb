```python
def preprocess_text(text):
    tokens = nltk.word_tokenize(text)
    stop_words = set(nltk.corpus.stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
    return ' '.join(filtered_tokens)


new['Top1'] = new['Top1'].apply(preprocess_text)
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(new['Top1'])
```

```python
dense_matrix = X.todense() 
```

```python

doc_index = 0
tfidf_scores = X[doc_index]

feature_names = vectorizer.get_feature_names_out()

non_zero_indices = tfidf_scores.nonzero()[1]
non_zero_scores = tfidf_scores.data
import pandas as pd
df_tfidf = pd.DataFrame({
    'term': [feature_names[i] for i in non_zero_indices],
    'tfidf_score': non_zero_scores
})

print(df_tfidf)
```
