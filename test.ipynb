```python
from sklearn.cluster import KMeans

data = df[['Age','Fare']]


liners = []


for i in range(1,11):
    kamens = KMeans(n_clusters = i)
    kamens.fit(data)
    liners.append(kamens.inertia_)

    
    
    
plt.plot(range(1,11),liners,marker = 'o') 
plt.show()   

```
